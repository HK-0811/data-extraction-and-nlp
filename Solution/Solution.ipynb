```python
import os
import pandas as pd
import requests
from bs4 import BeautifulSoup
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
import re
import string
```

```python
# Define a function to extract the article text from a URL
def extract_article_text(url):
    try:
        response = requests.get(url)
        print(f"Response status code: {response.status_code}") 
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')
            title = ""
            article_text = ""
            
            title = soup.find("h1", class_="entry-title")
            title = title.get_text() if title else None
            
            article_text = soup.find("div",class_ = "td-post-content tagdiv-type")
            article_text = article_text.get_text() if article_text else None
        
            return title,article_text
        
    except Exception as e:
        print(f"Error extracting data from {url}: {e}")
    return None, None
```

```python
# Folder to save extracted files
output_folder = ('extracted_articles')
os.makedirs(output_folder, exist_ok=True)
```

```python
#To iterate Through url
for index, row in data.iterrows():
    url_id = row['URL_ID']
    url = row['URL']
    print(url)
    title,article_text = extract_article_text(url)
    if title is not None and article_text is not None:
        filename = os.path.join(output_folder, f"{url_id}.txt")
        with open(filename, 'w', encoding='utf-8') as file:
            file.write(f"{title}\n\n{article_text}")
        print(f"Saved {filename}")
    else:
        print(f"Skipped {url_id}: Unable to extract data from {url}")
        
        

print("Extraction and saving complete.")
```

```python
extracted_text_folder = r"C:\Users\Himanshu Kotkar\Desktop\extracted_articles"
computed_variables = {
    "URL_ID": [],
    "PositiveScore": [],
    "NegativeScore": [],
    "PolarityScore": [],
    "SubjectivityScore": [],
    "AverageSentenceLength": [],
    "Percentage of ComplexWords": [],
    "FogIndex": [],
    "Average Number of Words Per Sentence":[],
    "Complex Word Count":[],
    "Word Count":[],
    "Syllable Count Per Word":[],
    "Personal Pronouns" : [],
    "Average Word Length":[]
}

analyzer = SentimentIntensityAnalyzer()
positive_dictionary = [] 
negative_dictionary = []  
```

```python
positive_words_file = r"C:\Users\Himanshu Kotkar\Desktop\MasterDictionary\positive-words.txt"
with open(positive_words_file, 'r') as file:
    positive_words = [line.strip() for line in file]

positive_dictionary.extend(positive_words)


negative_words_file = r"C:\Users\Himanshu Kotkar\Desktop\MasterDictionary\negative-words.txt"
with open(negative_words_file, 'r') as file:
    negative_words = [line.strip() for line in file]

negative_dictionary.extend(negative_words)
```

```python
def syllable_count(word):
    count = 0
    vowels = "aeiou"

    if word[0] in vowels:
        count += 1

    for i in range(1, len(word)):
        if word[i] in vowels and word[i - 1] not in vowels:
            count += 1

    if word.endswith("es") or word.endswith("ed"):
        count -= 1

    return max(1, count)
```

```python
# Define a regex pattern to match personal pronouns and exclude "US"
personal_pronoun_pattern = r'\b(I|we|my|ours|us)\b'
# Compile the regex pattern
personal_pronoun_regex = re.compile(personal_pronoun_pattern, re.IGNORECASE)
```

```python
extracted_text_folder = r"C:\Users\Himanshu Kotkar\Desktop\extracted_articles"
complex_word_count = 0 
stopword_nltk_set = set(stopwords.words('english'))

for filename in os.listdir(extracted_text_folder):
    if filename.endswith(".txt"):
        with open(os.path.join(extracted_text_folder, filename), 'r', encoding='utf-8') as file:
            article_text = file.read()

        # Tokenize the article text
        words = nltk.word_tokenize(article_text)
        sentences = sent_tokenize(article_text)

        # Remove stopwords
        cleaned_words = [word for word in words if word.lower() not in stopword_set]

        # Calculate Positive Score
        positive_scores = sum(1 for word in cleaned_words if word.lower() in positive_dictionary)

        # Calculate Negative Score
        negative_score = -sum(1 for word in cleaned_words if word.lower() in negative_dictionary)
        negative_scores = negative_score * -1

        # Calculate Polarity Score
        polarity_scores = (positive_score - negative_score) / ((positive_score + negative_score) + 0.000001)

        # Calculate Subjectivity Score
        subjectivity_scores = abs  ((positive_score + negative_score) / ((len(cleaned_words) + 0.000001)))
        
        # Calculate the number of syllables for each word
        syllables = [syllable_count(word) for word in cleaned_words]
        
        # Count the number of words with more than two syllables as complex words
        complex_word_count = sum(1 for s in syllables if s > 2)
        
        # Calculate the Average Sentence Length
        num_sentences = len(sentences)
        num_words = len(cleaned_words)
        average_sentence_length = num_words / num_sentences

        # Calculate the Percentage of Complex Words
        percentage_complex_words = (complex_word_count / num_words)*100

        # Calculate the Fog Index
        fog_index = 0.4 * (average_sentence_length + percentage_complex_words)
        
        # Calculate the Average Number of Words Per Sentence
        average_words_per_sentence = num_words / num_sentences
        
        # Calculate the Word Count and remove stopwordds using nltk
        cleaned_words_nltk = [
            word.lower() 
            for word in words if word.lower() not in stopword_nltk_set 
            and word not in string.punctuation
        ]
        num_nltk_words = len(cleaned_words_nltk)
        word_count = num_nltk_words
        
        # Calculate the Syllable Count Per Word
        syllable_count_per_word = sum(syllables) / len(syllables) if len(syllables) > 0 else 0
        
        # Count personal pronouns (excluding "US") using regex
        personal_pronouns_count = len(personal_pronoun_regex.findall(article_text))
        
        # Calculate the Average Word Length
        average_word_length = sum(len(word) for word in cleaned_words) / len(cleaned_words) if len(cleaned_words) > 0 else 0


        # Append the computed variables to the dictionary
        computed_variables["URL_ID"].append(filename[:-4])  # Remove the '.txt' extension
        computed_variables["PositiveScore"].append(positive_scores)
        computed_variables["NegativeScore"].append(negative_scores)
        computed_variables["PolarityScore"].append(polarity_scores)
        computed_variables["SubjectivityScore"].append(subjectivity_scores)
        computed_variables["AverageSentenceLength"].append(average_sentence_length)
        computed_variables["Percentage of ComplexWords"].append(percentage_complex_words)
        computed_variables["FogIndex"].append(fog_index)
        computed_variables["Average Number of Words Per Sentence"].append( average_words_per_sentence)
        computed_variables["Complex Word Count"].append(complex_word_count)
        computed_variables["Word Count"].append(word_count)
        computed_variables["Syllable Count Per Word"].append(syllable_count_per_word)
        computed_variables["Personal Pronouns"].append(personal_pronouns_count)
        computed_variables["Average Word Length"].append( average_word_length)              
```

```python
computed_var_df = pd.DataFrame(computed_variables)
```

```python
computed_var_df.to_csv("computed_variabless.csv",index=False)
print("Variable computation complete.")
```
